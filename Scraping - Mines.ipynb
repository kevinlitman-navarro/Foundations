```python
driver = webdriver.Chrome()
```

```python
driver.get('https://arlweb.msha.gov/drs/drshome.htm')
```

```python
# Need to uncheck this box to ensure that we are including the abandoned mines
checkbox = driver.find_element_by_name('Abandoned')
checkbox.click()
```

```python
# Now to fill the search field

operator_name = driver.find_element_by_name('OperSearch')
operator_name.send_keys('dirt')

# And to click the search button

button = driver.find_element_by_xpath('//*[@id="content"]/form[1]/table/tbody/tr[7]/td[3]/input[1]')
button.click()
```

### Hopefully you know that each `tr` is supposed to be a row of your data. What is the index of the first row element that is actually a result?

> - *Tip: `.text` will help you here.*
> - *Tip: You aren't interesting in annotations or anything, just mines and where they are from*
> - *Tip: Using `print("-----")` will help you keep track of different rows*

```python
import pandas as pd
```

```python
data = driver.find_elements_by_tag_name('tr')

# After looking at the full list, I know I can slice the beginning (index)7 items off to get to what I want
# Also going to slice off the final entry 

for dat in data[7:-1]:
    
    print("---------")
    print(dat.text)

# Now that I kknow it's working, i'm gonna make a new variable

dirt_mines = data[7:-1]
```

### Loop through each operator result, printing its name

> - *Tip: If you have a list called `animals`, `animals[2:]` will skip the first two and start with the third.*
> - *Tip: You can use list slicing or an `if` statement to skip the non-data row(s). List slicing is probably easier, even if you aren't comfortable with it.*
> - *Tip: or honestly you can use `try` and `except` if you know how it works.*
> - *Tip: Once you have the "right" rows of data, you're going to be looking for a certain tag inside*
> - *Tip: Sometimes you can't say "give me this class," and instead you have to say "give me all of the `div` elements, and then give me the third one."*

```python
for mine in dirt_mines:
    
    # Each value is stored in a 'td' tag in order, so the 3rd 'td' tag in each 'tr' contains the operator name
    title = mine.find_elements_by_tag_name('td')[2].text
    print(title)
```

### Loop through each operator result, printing its ID

There should be ONE code per row, and NO empty rows between them.

```python
for mine in dirt_mines:
    
    # Each value is stored in a 'td' tag in order, so the 1st 'td' tag in each 'tr' contains the ID
    ID = mine.find_elements_by_tag_name('td')[0]
    print(ID.text)
```

## Saving the results

### Loop through each `tr` to create a list of dictionaries

Each dictionary must contain

- Operator ID
- Operator name
- Mine name
- State
- Mine type
- Coal or metal
- Status
- Commodity

Create a new dictionary for each row.

> - *Tip: Start with an empty dictionary, then add the keys one at a time like we did during class*
> - *Tip: You might want to save all of the cells in a variable, then use indexes to get the second, third, fourth, etc.*
> - *Tip: I know you already skipped a bunch of rows already, but one of them still might be bad! Which one is it? How can you skip it? You might need to slice out some of the end of your list, too. Use `print` to help you debug, or just look at the page closely.*
> - *Tip: Or, if you did the other homework already, `try` / `except` is also an option*

```python
rows = []

for mine in dirt_mines:
    row = {}
    
    #print("---------")
    
    # Add operator ID to dict
    
    ID = mine.find_elements_by_tag_name('td')[0].text
    #print(ID)
    row['ID'] = ID
    
    # Add operator name to dict
    
    operator_name = mine.find_elements_by_tag_name('td')[2].text
    #print(operator_name)
    row['operator_name'] = operator_name
    
    # Add mine name to dict
    
    mine_name = mine.find_elements_by_tag_name('td')[3].text
    #print(mine_name)
    row['mine_name'] = mine_name
    
    # Add state to dict
    
    state = mine.find_elements_by_tag_name('td')[1].text
    #print(state)
    row['state'] = state
    
    # Add mine type to dict
    
    mine_type = mine.find_elements_by_tag_name('td')[4].text
    #print(mine_type)
    row['mine_type'] = mine_type
    
    # Add coal or metal to dict
    
    coal_or_metal = mine.find_elements_by_tag_name('td')[5].text
    
    #Going to clean it up a bit
    if coal_or_metal == "M ":
        coal_or_metal = "Metal"
    elif coal_or_metal == "C ":
        coal_or_metal = "Coal"
        
    #print(coal_or_metal)
    row['coal_or_metal'] = coal_or_metal
    
    # Add status to dict
    
    status = mine.find_elements_by_tag_name('td')[6].text
    #print(status)
    row['status'] = status
    
    # Add commodity to dict
    
    commodity = mine.find_elements_by_tag_name('td')[7].text
    #print(commodity)
    row['commodity'] = commodity
    
    #print(row)
    rows.append(row)
    
print(rows[0])
```

### Save that to a CSV named `dirt-operators.csv`

```python
df = pd.DataFrame(rows)

df

df.to_csv("dirt-operators.csv", index=False)
```
